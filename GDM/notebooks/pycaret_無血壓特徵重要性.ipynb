```python
import pandas as pd
import seaborn as sns
from pycaret.classification import setup
from pycaret.classification import compare_models
from pycaret.classification import create_model
import time


# 讀取資料集
data = pd.read_csv('1.csv')

# 設置 PyCaret 的設定
setup(data, target='GDM')  

# 記錄開始時間
start_time = time.time()

# 比較不同模型的性能
best_model = compare_models()

print(type(best_model))

# 記錄結束時間
end_time = time.time()

# 計算花費時間
elapsed_time = end_time - start_time

# 顯示花費時間
print("訓練時間：{:.2f} 秒".format(elapsed_time))
```

```python
# 建立一個決策樹分類器模型
model = create_model('ada')

# 檢查模型的特徵重要性
if hasattr(model, 'feature_importances_'):
    feature_importances = model.feature_importances_
    feature_names = data.columns
    # 顯示特徵名稱及特徵重要性
    for i, feature_importance in enumerate(feature_importances):
        feature_name = feature_names[i]
        print("特徵名稱: {}, 特徵重要性: {}".format(feature_name, feature_importance))
else:
    print("此模型不提供特徵重要性。")
    
# 選擇重要性高的特徵
if hasattr(model, 'feature_importances_'):
    selected_features = [f for f, importance in zip(data.columns, model.feature_importances_) if importance > 0.01]
    print("重要性高於閾值的特徵:", selected_features)
else:
    print("此模型不提供特徵重要性。")
```
